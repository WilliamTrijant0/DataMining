```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('dataprepocessing.csv')
df
```

```python
df_baru = df.loc[:,['jenis','jumlah_penumpang','jenis_kepadatan','Prediksi']]
df_baru
```

```python
df_baru.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean', missing_values=np.nan)
imputer = imputer.fit(df_baru[['jumlah_penumpang']])
df_baru['jumlah_penumpang'] = imputer.transform(df_baru[['jumlah_penumpang']])
df_baru
```

```python
df_baru.isnull().sum()
```

```python
nama_kepadatan = dict(zip(df_baru.Prediksi.unique(), df_baru.jenis_kepadatan.unique()))
nama_kepadatan
```

```python
from sklearn.model_selection import train_test_split
x = df_baru[['jumlah_penumpang']]
y = df_baru['Prediksi']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=1)
```
